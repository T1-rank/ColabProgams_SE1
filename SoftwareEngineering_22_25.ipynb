```python
import matplotlib.pyplot as plt
import numpy as np

def quadratic_model(x, a, b, c):
    return a * x**2 + b * x + c

def generate_weather_data(start, end, num_points, a, b, c, noise_factor=10):
    x = np.linspace(start, end, num_points)
    y = quadratic_model(x, a, b, c) + np.random.normal(0, noise_factor, num_points)
    return x, y

def plot_weather_data(x, y, title):
    plt.scatter(x, y, label='Weather Data')
    plt.title(title)
    plt.xlabel('Time')
    plt.ylabel('Temperature')
    plt.legend()
    plt.show()

# Example parameters for the quadratic model
a = 0.5
b = -5
c = 20

# Generate weather data
start_time = 0
end_time = 10
num_points = 100

x_data, y_data = generate_weather_data(start_time, end_time, num_points, a, b, c)

# Waterfall Model (example title)
plot_weather_data(x_data, y_data, 'Waterfall Model: Weather Data')

# Display the final temperature value
final_temperature = y_data[-1]
print(f"Final Temperature for Waterfall Model (Example Parameters): {final_temperature}")
```

```python
import matplotlib.pyplot as plt
import numpy as np

def quadratic_model(x, a, b, c):
    return a * x**2 + b * x + c

def generate_weather_data(start, end, num_points, a, b, c, noise_factor=10):
    x = np.linspace(start, end, num_points)
    y = quadratic_model(x, a, b, c) + np.random.normal(0, noise_factor, num_points)
    return x, y

def plot_weather_data(x, y, title):
    plt.figure()
    plt.scatter(x, y, label="Weather Data")
    plt.title(title)
    plt.xlabel("Time")
    plt.ylabel("Temperature")
    plt.legend()
    plt.show()

# ---------------- INITIAL PARAMETERS ----------------
a_initial = 0.5
b_initial = -5
c_initial = 20

start_time = 0
end_time = 10
num_points = 100

# Generate initial weather data
x_data, y_initial = generate_weather_data(
    start_time, end_time, num_points,
    a_initial, b_initial, c_initial
)

# ---------------- INITIAL GRAPH ----------------
plot_weather_data(
    x_data, y_initial,
    "Iterative Model: Initial Weather Data"
)

# ---------------- ITERATIVE DEVELOPMENT ----------------
num_iterations = 3

for iteration in range(1, num_iterations + 1):
    print(f"\nIteration {iteration}: Gather feedback and refine the model.")

    a_iterative = float(input("Enter new value for 'a': "))
    b_iterative = float(input("Enter new value for 'b': "))
    c_iterative = float(input("Enter new value for 'c': "))

    # Generate updated weather data (no noise for clarity)
    y_iterative = quadratic_model(
        x_data,
        a_iterative,
        b_iterative,
        c_iterative
    )

    # Plot each iteration separately
    plot_weather_data(
        x_data,
        y_iterative,
        f"Iterative Model: Iteration {iteration} Weather Data"
    )

    if iteration == num_iterations:
        print(f"\nTemperature values for Iteration {iteration}:\n{y_iterative}")

# ---------------- FINAL GRAPH ----------------
plot_weather_data(
    x_data,
    y_iterative,
    "Iterative Model: Final Weather Data"
)
```

```python
import matplotlib.pyplot as plt
import numpy as np

def quadratic_model(x, a, b, c):
    return a * x**2 + b * x + c

def generate_weather_data(start, end, num_points, a, b, c, noise_factor=10):
    x = np.linspace(start, end, num_points)
    y = quadratic_model(x, a, b, c) + np.random.normal(0, noise_factor, num_points)
    return x, y

def plot_weather_data(x, y, title):
    plt.figure()
    plt.scatter(x, y, label="Weather Data")
    plt.title(title)
    plt.xlabel("Time")
    plt.ylabel("Temperature")
    plt.legend()
    plt.show()

# ---------------- INITIAL PARAMETERS ----------------
a_agile = 0.5
b_agile = -5
c_agile = 20

start_time = 0
end_time = 10
num_points = 100

# Generate initial weather data
x_data, y_agile = generate_weather_data(
    start_time, end_time, num_points,
    a_agile, b_agile, c_agile
)

# ---------------- INITIAL GRAPH ----------------
plot_weather_data(
    x_data, y_agile,
    "Agile Model: Initial Weather Data"
)

# ---------------- AGILE DEVELOPMENT (SPRINTS) ----------------
num_sprints = 3

for sprint in range(1, num_sprints + 1):
    print(f"\nSprint {sprint}: Gather feedback and refine the model.")

    a_agile += float(input("Enter incremental change for 'a': "))
    b_agile += float(input("Enter incremental change for 'b': "))
    c_agile += float(input("Enter incremental change for 'c': "))

    # Generate updated weather data
    y_agile = quadratic_model(
        x_data,
        a_agile,
        b_agile,
        c_agile
    )

    # Plot sprint-wise graph (SEPARATE)
    plot_weather_data(
        x_data, y_agile,
        f"Agile Model: Sprint {sprint} Weather Data"
    )

# ---------------- FINAL GRAPH ----------------
plot_weather_data(
    x_data, y_agile,
    "Agile Model: Final Weather Data"
)

final_temperature_agile = y_agile[-1]
print(f"\nFinal Temperature for Agile Model: {final_temperature_agile}")
```
