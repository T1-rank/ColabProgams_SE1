```python
# Version 3: Read From File (Single Input)

# Save the file
with open("single_input.txt", "w") as f:
    f.write("42")

# Read single input from file
try:
    with open("single_input.txt", "r") as f:
        num = int(f.read().strip())

    result = num * 2
    print(f"Single Input Read from File: {num} * 2 = {result}")

except FileNotFoundError:
    print("File not found!")

except ValueError:
    print("Error in reading or converting the input!")
```

```python
# Version 4: Read From File (Multiple Inputs)

# Save file with multiple inputs
with open("multiple_inputs.txt", "w") as f:
    f.write("3\n5\n7")

# Read multiple inputs from file
try:
    with open("multiple_inputs.txt", "r") as f:
        numbers = [int(line.strip()) for line in f.readlines()]

    result = sum(numbers)
    print(f"Multiple Inputs Read from File: {numbers}, Sum = {result}")

except FileNotFoundError:
    print("File not found!")

except ValueError:
    print("Error in reading or converting the inputs!")
```

```python
import matplotlib.pyplot as plt
import numpy as np

def quadratic_model(x, a, b, c):
    return a * x**2 + b * x + c

def generate_weather_data(start, end, num_points, a, b, c, noise_factor=10):
    x = np.linspace(start, end, num_points)
    y = quadratic_model(x, a, b, c) + np.random.normal(0, noise_factor, num_points)
    return x, y

def plot_weather_data(x, y, title):
    plt.scatter(x, y, label='Weather Data')
    plt.title(title)
    plt.xlabel('Time')
    plt.ylabel('Temperature')
    plt.legend()
    plt.savefig("weather_plot.png")  # ðŸ”¹ Saves graph for GitHub
    plt.show()

# Example parameters for the quadratic model
a = 1.5
b = -8
c = 30

# Generate weather data
start_time = 0
end_time = 10
num_points = 100

x_data, y_data = generate_weather_data(start_time, end_time, num_points, a, b, c)

# Waterfall Model
plot_weather_data(x_data, y_data, "Waterfall Model: Weather Data")

# Display final temperature
final_temperature = y_data[-1]
print(f"Final Temperature for Waterfall Model (Example Parameters): {final_temperature}")
```
